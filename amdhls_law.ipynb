```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def amdahl_speedup(frac_enhanced, speedup_enhanced):
    net_speedup = 1 / ((1 - frac_enhanced) + (frac_enhanced / speedup_enhanced))
    return net_speedup
```

```python
percent = np.array(range(0, 101, 10))
```

```python
improvement_list = []
for i in percent:
    improvement_list.append(amdahl_speedup(i/100, 8))
    print(f"Speedup for {i}% enhanced: {amdahl_speedup(i/100, 8)}")
```

```python
plt.figure()
plt.xlabel("Percent Vectorized")
plt.ylabel("Net Speedup")
plt.title("Speedup based on Amdahl's Law")
plt.grid()
plt.xticks(percent)
plt.legend()
plt.plot(percent, improvement_list, marker="o", label="Speedup", linestyle="--")
plt.show()
```
